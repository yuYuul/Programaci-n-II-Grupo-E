```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("pacientePeso.txt", sep="|", header=None,
                 names=["Nombre","Fecha","Edad","Grupo","Seguro","Centro","Peso","Sexo"])

print(" Datos registrados en el archivo:")
display(df)

peso_max = df["Peso"].max()
peso_min = df["Peso"].min()
peso_prom = df["Peso"].mean()

print("\n Indicadores de peso:")
print("Peso máximo:", peso_max, "kg")
print("Peso mínimo:", peso_min, "kg")
print("Peso promedio:", round(peso_prom,2), "kg")

print("\n Pacientes Masculinos:")
display(df[df["Sexo"] == "M"])

print("\n Pacientes Femeninos:")
display(df[df["Sexo"] == "F"])

plt.figure(figsize=(6,4))
df.groupby("Sexo")["Peso"].mean().plot(kind="bar", color=["blue","pink"])
plt.title("Comparación de Peso Promedio por Sexo")
plt.xlabel("Sexo")
plt.ylabel("Peso promedio (kg)")
plt.xticks(rotation=0)
plt.show()

```
