```
# Colab-ready: análisis de doctores
import pandas as pd
import matplotlib.pyplot as plt

# from google.colab import files
# uploaded = files.upload()  # sube doctoresRegistro.txt

# Leer archivo (si ya está en el entorno)
df = pd.read_csv('doctoresRegistro.txt', sep='|', header=None)
df.columns = ["Nombre","Especialidad","AñosExperiencia","Sexo","Hospital"]
df["AñosExperiencia"] = pd.to_numeric(df["AñosExperiencia"], errors='coerce')

# Mostrar
display(df)

# Estadísticas
print("Promedio de experiencia:", df["AñosExperiencia"].mean())
print("Máximo de experiencia:", df["AñosExperiencia"].max())
print("Mínimo de experiencia:", df["AñosExperiencia"].min())

# Filtrar por especialidad
print("Doctores de cardiologia:")
display(df[df["Especialidad"].str.lower()=="cardiologia"])
print("Doctores de pediatria:")
display(df[df["Especialidad"].str.lower()=="pediatria"])
print("Doctores de neurologia:")
display(df[df["Especialidad"].str.lower()=="neurologia"])

# Filtrar por sexo femenino
print("Doctores femeninos:")
display(df[df["Sexo"].str.lower()=="femenino"])

# Gráficos
df['Especialidad'].value_counts().plot(kind='bar')
plt.title('Cantidad de doctores por especialidad')
plt.xlabel('Especialidad')
plt.ylabel('Cantidad')
plt.show()

df['Sexo'].value_counts().plot(kind='pie', autopct='%1.1f%%')
plt.title('Distribución por sexo')
plt.show()
```
